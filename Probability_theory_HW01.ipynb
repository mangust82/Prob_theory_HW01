```python
#Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. 
#б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.


from IPython.display import display
# display(Math(r'F(k) = \int_{-\infty}^{\infty} f(x) e^{2\pi i k} dx'))


from numpy import math
print('A)')
print('Вероятность будет вычиcляться как P = N/M, где N положительное число исходов, M общее число исходов:')
print('Число возможных позитивных исходов:')
display(Math(r'N = \frac{n!}{k!(n-k)!}'))
Na = math.factorial(13)/(math.factorial(4)*math.factorial(13-4))
print(f'Найдем число исходов N, в колоде из 52 карт 13 крести, тогда количество исходов когда мы достанем 4 карты: {Na}')
M = math.factorial(52)/(math.factorial(4)*math.factorial(52-4))
print(f'Найдем общее число исходов M, в колоде из 52 при выборе 4 карт {M}')
Pa = Na/M
print(f'Вероятность достать все крести: {Pa}\n')

print('Б)')
K = math.factorial(48)/(math.factorial(4)*math.factorial(48-4))
print(f'Найдем число исходов K когда у нас не будет ни одного туза: {K}')
Nb = M - K
print(f'Тогда число исходов когда есть хотя бы один туз Nb = M - K: {Nb}')
Pb = Nb/M
print(f'вероятность, что среди 4-х карт окажется хотя бы один туз Pb: {Pb}')
```

```python
# На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. 
# Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, 
# откроет дверь с первой попытки?

M = math.factorial(10)/(math.factorial(3)*math.factorial(10-3))
print(f'Найдем возможное количество комбинаций M, при количестве цифр 10: {M}')
P = 1/M
print(f'Поскольку поскольку положительный исход только один N = 1, то вероятность открыть дверь с первой попытки: {P}')
```

```python
# В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. 
# Какова вероятность того, что все извлеченные детали окрашены?
print('Способ 1')
print('Найдем общую вероятность как произведение вероятностей достать каждый раз окрашенный шар Pk = P1*P2*P3')
print('где P1 = 9/15, P2 = 8/14, P3 = 7/13')
print(f'Pk = {9/15 * 8/14 * 7/13}\n')

print('Способ 2')
N = math.factorial(9)/(math.factorial(3)*math.factorial(9-3))
print(f'Найдем позитивное число исходов N: {N}')
M = math.factorial(15)/(math.factorial(3)*math.factorial(15-3))
print(f'Найдем общее число исходов M: {M}')
Pk = N/M
print(f'Вероятность достать все крашенные шары Pk={Pk}\n')

```
